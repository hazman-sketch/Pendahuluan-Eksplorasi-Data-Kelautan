# Groupby
Groupby bisa digunakan untuk merata-ratakan nilai parameter yang kita inginkan berdasarkan dimensi yang kita inginkan. Misalkan terdapat sekumpulan data tahun 2013, didalamnya terdapat data perbulan, dimana dalam satu bulan itu merupakan data harian dengan rentang data per 6 jam, kita akan merata-ratakan data setiap bulannya. Kita bisa menggunakan groupby untuk hal ini.

```python
# groupby data menjadi rata-rata bulanan
dataset.air.groupby('time.month')
```

```python
#average setiap bulannya
monthly_climatology = dataset.air.groupby('time.month').mean('time')
monthly_climatology
```

```python
monthly_climatology.mean(['lat', 'lon']).plot()
```

Grafik diatas merupakan seasonal variation iklim di jepang, dimana kita dapatkan rata-rata air temperature bulanan yang kita rata-ratakan dari data tahunan tadi

```python
# Contoh perbandingan iklim asia

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=[12,4])

monthly_climatology.sel(month=12).plot(ax=ax1, vmin=240, vmax=300, extend='both', 
                                      add_colorbar=False, cmap='turbo')
im=monthly_climatology.sel(month=6).plot(ax=ax2, vmin=240, vmax=300, extend='both', 
                                         add_colorbar=False, cmap='turbo')

cbar=fig.colorbar(im, extend='both', ax=(ax1,ax2))
```

Pada latitude yang lebih besar, selama periode summer (Juni hingga Agustus) temperatur lebih hangat, ditunjukan dengan gambar yang sebelah kanan

```python
# groupby dengan periode waktu tahunan
yearly=dataset.air.groupby('time.year').mean('time')
yearly
```

```python
temperature_difference = yearly.sel(year=2014) - yearly.sel(year=2013)

temperature_difference.plot(cbar_kwargs={'label':'Temperature Difference'})
```

Diatas merupakan contoh analisis perbedaan temperatur 2014 dengan 2013

# Resampling
Resampling ini merupakan proses rata-rata dimana ketika kita memiliki data per 6 jam menjadi harian, dan lain-lain  

```python
dataset.resample(time='D')
```

```python
# Data per 6 jam menjadi harian

daily=dataset.resample(time='D').mean('time')
daily
```

```python
# Data per 6 Jam kita jadikan bulanan

monthly=dataset.resample(time='MS').mean('time')
monthly

#'MS' menunjukan data tanggal yang digunakan adalah tangal awal bulan
```

```python
# plot to see the difference
plt.figure(dpi=300, figsize=[9,3])

daily.air.mean(['lat','lon']).plot(label='daily')
monthly.air.mean(['lat','lon']).plot(marker='o', label='monthly')

# turn on the legend
plt.legend()
```

Diatas merupakan gambaran dari perbandingan data resampling harian (biru) dengan bulanan (orange).

# Computation
Dalam bab ini diajarkan proses kalkulasi dari data yang kita miliki seperti convert data suhu dari kelvin ke celcius, dan masih banyak lagi.

```python
ds_jan2013
```

```python
# convert dari kelvin ke ˚C

ds_jan2013_celcius = ds_jan2013 - 273.15
ds_jan2013_celcius
```

```python
# berikut contoh plot perbedaan dari data yang masih kelvin dan sudah menjadi celcius
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=[12,4])

ds_jan2013.mean('time').plot(ax=ax1, cmap='turbo', cbar_kwargs={'label':'Air Temperature [K]'})
ds_jan2013_celcius.mean('time').plot(ax=ax2, cmap='turbo', cbar_kwargs={'label':'Air Temperature [˚C]'})

ax1.set_title('Air Temperature in Kelvin (Jan 2013)')
ax2.set_title('Air Temperature in Celcius (Jan 2013)')
```

```python
# Kalkulasi antara dua data arrays

first_jan = ds_jan2013.isel(time=0)
last_jan = ds_jan2013.isel(time=-1)

first_minus_last = first_jan - last_jan
```

```python
first_minus_last.plot()

plt.title('1 Jan 2013 minus 31 Jan 2013')
```

# Plotting
Aturan plotting xarray : Kegiatan plotting harus dilakukan dengan 2 (atau kurang) dimensi. Tidak bisa dilakukan plotting jika dimensinya masih 3. Bisa kita lihat pada data atribut atau keterangan dari dataset yang kita panggil. Ketika nama dimensi masih bold itu menunjukkan banyaknya dimensi yang masih aktif atau belum kita tentukan.

```python
ds_jan2013
```

Note : Pada coordinat lat, lon, dan time masih bercetak tebal hal tersebut masih menunjukan dimensi data masih terdapat tiga dimensi. Sehingga untuk plotting xarray kita gunakan 2 dimensi saja.

```python
example = ds_jan2013.isel(time=15)
example
```

Note : Pada keterangan coordinat : lat dan lon masih bercetak tebal, namun dimensi time sudah tidak tebal. Menunjukan 2 dimensi dan dapat kita plot 

```python
example.plot()
```

```python
#Jika kita hanya gunakan 1 dimension saja, akan otomatis menghasilkan gambar grafik ketika kita plotting

example.sel(lat=37, method='nearest')
```

# Contoh plotting
Berikut adalah sala satu contoh plotting, anda bisa gunakan untuk kegiatan plotting di kesempatan lain.

```python
fig, ax1 = plt.subplots(dpi=300, # higher dpi higher figure quality
                       figsize=[9,3]) # the width and height of the figure, respectively

example.plot(ax=ax1, # assign the axis
             cmap='Spectral', # change the colormap here
             vmin=240, # set the lower value of the colorbar
             vmax=300, # set the upper value of the colorbar
            cbar_kwargs={'shrink':0.7, # shrink the colorbar
                        'label':'colorbar label here'}) 

ax1.set_title('Write the title here')
```
